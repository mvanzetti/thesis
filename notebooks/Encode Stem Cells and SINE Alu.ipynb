## Stem Cells

```python
import seaborn as sns
%matplotlib inline
import pandas as pd
```

```python
stem_df = pd.read_hdf('../storage/encode_repeatmasker_stats.hdf', 'reldist_hg19_DNase_H3K27ac_stem_cell_SINE_Alu')
```

```python
stem_df.T
```

```python
stem_df = stem_df.query("ovlp_fraction > 0")
```

```python
sns.set(style="whitegrid")
sns.set_context("poster", font_scale=0.8, rc={"lines.linewidth": 0.8})

g = sns.FacetGrid(stem_df, col="biosample_name", hue="ovlp_encyclopedia", col_wrap=2, size=8)
g.map(sns.pointplot, "reldist", "ovlp_fraction", capsize=.2, size=6, aspect=1)
g.add_legend();
```

```python
stem_ovlp_df = pd.read_hdf('../storage/encode_repeatmasker_stats.hdf', 'overlap_hg19_DNase_H3K27ac_stem_cell_SINE_Alu')
```

```python
stem_ovlp_df.T
```
