## ENCODE APIs

```python
import json
import urllib.request
import requests
from pandas.io.json import json_normalize
```

```python
headers = {'accept': 'application/json'}
main_url = "https://www.encodeproject.org"
search_url = main_url + "/search/?type=Annotation&annotation_type=enhancer-like+regions&frame=object&limit=all"

response = requests.get(search_url, headers=headers)
response_json_dict = response.json()

graph = response_json_dict["@graph"]
```

```python
row_df = json_normalize(graph[0])
assembly_list = row_df.get_value(0, "assembly")
if len(assembly_list) > 0:
    row_df['assembly'] = assembly_list[0]
row_df
```

### Overlaps using pybedtools

```python
# find overlaps using pybedtools
adrenal_gland_dbSUPER_bed_filepath = dbSUPER_filepath + 'adrenal_gland_dbSUPER_sample.bed'
adrenal_gland_dbSUPER_bed = BedTool(adrenal_gland_dbSUPER_bed_filepath)

adrenal_gland_ENCODE_bed_filepath = dbSUPER_filepath + 'adrenal_gland_ENCODE_sample.bed'
adrenal_gland_ENCODE_bed = BedTool(adrenal_gland_ENCODE_bed_filepath)
```

```python
# wa = mantains the original A feature for A.intersect(B), u = unique, 
# loj = Left outer join: report features in A with and without overlaps
# f = by default, bedtools intersect will report an overlap between A and B 
# so long as there is at least one base pair is overlapping. 
# To restrict reported overlaps between A and B to cases where the feature in B overlaps at least X% (e.g. 50%) 
# of the A feature

#the exact list obtained from the upload tool is obtainable with:
#ag_dbSUPER_intersect_ENCODE = adrenal_gland_dbSUPER_bed.intersect(adrenal_gland_ENCODE_bed, wa=True, f=0.1)

ag_dbSUPER_intersect_ENCODE = adrenal_gland_dbSUPER_bed.intersect(adrenal_gland_ENCODE_bed, wa=True, f=0.1)
ag_ENCODE_intersect_dbSUPER = adrenal_gland_ENCODE_bed.intersect(adrenal_gland_dbSUPER_bed, loj=True, f=0.1)

print(len(ag_dbSUPER_intersect_ENCODE), "dbSUPER.intersect(ENCODE) results")
print(len(ag_ENCODE_intersect_dbSUPER), "ENCODE.intersect(dbSUPER) results")
```
